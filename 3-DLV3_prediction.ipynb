```python
import tensorflow as tf
import pandas as pd
import numpy as np
import assist_function
```

# load model

```python
model = tf.keras.models.load_model('trained_models\DLV3_model.hdf5')
model.summary()
```

# load data

```python
current_model = [224, 224, 3]

X_pred = assist_function.load_and_resize_pred(
    "data/input_data/all_GSV_img_paths.txt", current_model)

print(X_pred.shape)
```

```python
pred_metadata = assist_function.retrieve_metadata('data/input_data/all_GSV_img_paths.txt')
pred_metadata.head(3)
```

# predict

```python
import time
st =  time.time()

pred_ts = model.predict(X_pred, batch_size=32)
pred_ts.shape

et =  time.time()
elapsed_time = et - st
print('inference time = ', elapsed_time,'seconds')
```

```python
import numpy as np
import pandas as pd 
import os
import assist_function
```

```python
test_path = "data\input_data\labelled_pair_test.txt"
test_img_paths, test_lbl_paths = assist_function.save_file_path_in_list(test_path)
```

```python
test_metadata = assist_function.retrieve_metadata(test_path)
test_metadata
```

# save test images in separate folder to upload

```python
from PIL import Image

for i, path1 in enumerate(test_img_paths):
    img = Image.open(path1)
    filename = f"{test_metadata.iloc[i, 0]}-{test_metadata.iloc[i, 1]}-{test_metadata.iloc[i, 2]}.jpg"
    save_path = os.path.join("images", "test_img_to_upload", "img", filename)
    img.save(save_path)
```

```python
# save test label images
for i, path1 in enumerate(test_lbl_paths):
    img = Image.open(path1)
    filename = f"{test_metadata.iloc[i, 0]}-{test_metadata.iloc[i, 1]}-{test_metadata.iloc[i, 2]}.jpg"
    save_path = os.path.join("images", "test_img_to_upload", "lbl", filename)
    img.save(save_path)
```

```python
# check if the img and lbl are identical
for i, zip_ in enumerate(zip(test_img_paths, test_lbl_paths)):
    img, lbl = zip_
    if os.path.basename(img) != os.path.basename(lbl):
        print(i)
```

# get onedrive embed data for each file

```python
# the df below shows that they have the same sequence as the test_evaluation df.
test_img_loc = "data\\test_embed_img_paths.txt"
test_lbl_loc = "data\\test_embed_lbl_paths.txt"
def retrieve_data(path):

    data, gt = assist_function.save_file_path_in_list(path)

    label_metadata = []
    for file_name in data:
        # Retrieve the file name
        file_name = os.path.basename(file_name)

        # Remove the file extension
        # Split the file name into its components using the '-' separator
        file_components = file_name.split('-')
        print(file_components)
        # Extract the latitude, longitude, and heading information from the file name
        lat = (file_components[0])
        lng = (file_components[1])
        heading = int(file_components[2].split('.')[0])

        label_metadata.append([lat, lng, heading])

    return pd.DataFrame(label_metadata, columns=['lat', 'lng', 'heading'])

dashboard_test_metadata =  retrieve_data(test_img_loc)
dashboard_test_metadata
```

```python
gt, _ = assist_function.save_file_path_in_list(test_lbl_loc)
metadata, img = assist_function.save_file_path_in_list(test_img_loc)
```

```python
# wait to add lbl url data
test_dashboard = pd.concat([dashboard_test_metadata, pd.Series(img), pd.Series(gt)], axis=1)
test_dashboard.dtypes
```

```python
test_dashboard[['lat','lng']] = test_dashboard[['lat','lng']].astype(float)
```

```python
test_dashboard
```

```python
# read the test_evaluation files and concat

def merge_embed_link_with_test_evaluation_df(csv_path):

    df = pd.read_csv(csv_path).drop('Unnamed: 0', axis=1)
    df_merged = pd.merge(df, test_dashboard, on=['lat','lng','heading'], how='inner')
    df_merged.rename(columns = {0: 'img_url', 1: 'lbl_url'}, inplace=True)

    return df_merged

DCNN_error_analysis = merge_embed_link_with_test_evaluation_df("data\output_data\DCNN_test_evaluation.csv")
DLV3_error_analysis = merge_embed_link_with_test_evaluation_df("data\output_data\DLV3_test_evaluation.csv")
PixSeg_error_analysis = merge_embed_link_with_test_evaluation_df("data\output_data\PixSeg_test_evaluation.csv")
```

```python
DCNN_error_analysis['model'] = 'DCNN'
DLV3_error_analysis['model'] = 'DLV3'
PixSeg_error_analysis['model'] = 'PixSeg'
```
